# Exercise 4.9

\begin{align*}
p(\ddot{\frown}) &= 0.01\\
p(\ddot{\smile}) &= 1 - 0.01
\end{align*}

By Bayes' rule the probability is
\begin{align*}
p(\ddot{\frown} | +) &= \dfrac{p(+ | \ddot{\frown})p(\ddot{\frown})}{p(+ | \ddot{\frown})p(\ddot{\frown}) + p(+ | \ddot{\smile})p(\ddot{\smile})} \\\\
&= \dfrac{(0.95)(0.01)}{(0.95)(0.01) + (1 - 0.90)(1 - 0.01)}.
\end{align*}

```r
posterior = (0.95*0.01)/(0.95*0.01 + (1-0.90)*(1-0.01))
print(posterior)
```

So $p(\ddot{\frown}|+)\approx 0.09$.

# DBDA Exercise 5.1

```r
new_prior = (0.99*0.001)/(0.99*0.001 + 0.05*(1-0.001))
print(new_prior)
```

After the first test comes out positive, we already calculated in the textbook that the probability they now have the disease is $\approx 0.019$, which now becomes our new prior for $p(\ddot{\frown})$.

Assuming that the test results are independent of each other where
$$p(-,+|\ddot{\frown}) = p(-|\ddot{\frown}) \cdot p(+|\ddot{\frown}),$$
Bayes' rule then becomes
\begin{align*}
p(\ddot{\frown}|-,+) &= \dfrac{p(-|\ddot{\frown}) p(\ddot{\frown}|+)}{\sum_{\theta^*} p(-|\theta^*) p(\theta^*|+)}\\\\
&= \dfrac{(1-0.99)(0.019)}{(1-0.99)(0.019) + (1-0.05)(1-0.019)}.
\end{align*}

```r
new_posterior = ((1-0.99)*new_prior)/((1-0.99)*new_prior + (1-0.05)*(1-new_prior))
print(new_posterior)
```

So $p(\ddot{\frown}|-,+) \approx 0.0002$.

# DBDA Exercise 5.3

## Part (a)

Bayes' rule gives
\begin{align*}
p(\ddot{\frown}|-) &= \dfrac{p(-|\ddot{\frown})p(\ddot{\frown})}{\sum_{\theta^*} p(-|\theta^*) p(\theta^*)}\\\\
&=\dfrac{(1-0.99)(0.001)}{(1-0.99)(0.001) + (1-0.05)(1-0.001)}.
\end{align*}

```r
new_prior = ((1-0.99)*0.001)/((1-0.99)*0.001 + (1-0.05)*(1-0.001))
print(new_prior)
```

So $p(\ddot{\frown}|-) \approx 0.000011$

## Part (b)

Now we can apply the next piece of information which is a subsequent positive test result. Since these tests are independent, we can write Bayes' rule as the following.
\begin{align*}
p(\ddot{\frown}|+,-) &= \dfrac{p(+|\ddot{\frown}) p(\ddot{\frown}|-)}{\sum_{\theta^*} p(+|\theta^*) p(\theta^*|-)}\\\\
&= \dfrac{(0.99)(0.000011)}{(0.99)(0.000011) + (0.05)(1-0.000011)}.
\end{align*}

```r
new_posterior = (0.99*new_prior)/(0.99*new_prior + 0.05*(1-new_prior))
print(new_posterior)
```

So $p(\ddot{\frown}|+,-) = p(\ddot{\frown}|-,+) \approx 0.0002$
