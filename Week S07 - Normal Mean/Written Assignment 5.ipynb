## Part (a)

We first calcualte $s'$ using the formula below.

\begin{align*}
\frac1{(s')^2} &= \frac1{s^2}+\frac{n}{\sigma^2}\\
&= \frac{1}{10^2} + \frac{10}{3^2}\\
&\approx 1.12\\\\
s' &= \sqrt{\frac1{1.12}}\\
&\approx 0.94
\end{align*}

The mean can be calculated using the formula below, which was rearranged to be in terms of $(s')^2$. We also calcualte $\bar{y} = 36.93$.

\begin{align*}
m' &= \frac{(s')^2}{s^2}\times m + \frac{n(s')^2}{\sigma^2}\times \bar{y}\\
&\approx \frac{0.89}{10^2}\times 30 + \frac{10\cdot0.89}{3^2}\times 36.93\\
&\approx 36.87
\end{align*}

The posterior distribution is then $normal(36.87, 0.94^2)$

## Part (b)

Since we are trying to test if the process icnreased the yield, our alternative hypothesis will be that the new mean is greater than 35. Our null will then be that the mean yield was unchanged or that the mean yield actually decreased.

$$H_0:\ \mu \le 35$$
$$H_1:\ \mu > 35$$

## Part (c)

We can calculate the z-score and calculate the p-value as follows:
\begin{align*}
P(\mu \le 35) &= P\left(z \le \frac{35 - 36.87}{0.944}\right)\\
&= P(z \le -1.98)\\
&= 0.02.
\end{align*}
Since the p-value is less than 0.05, we reject the null hypothesis. There is significant evidence to support the claim that the modified process increases the yield.
