## Part (a)

The distribution of $y$ is binomial, specifically $binom(n=30, p=\pi)$

## Part (b)

The frequentist estimate for $\pi$ is simply $\frac8{30} \approx 0.27$.

## Part (c)

A uniform prior is $beta(1,1)$. To update the prior according to the data we add the number of "successes" to $a$ and the number of "failures" to $b$. This gives a poster $g(\pi\mid y) = beta(9, 23)$.

## Part (d)

The Bayesian estimate of $\pi$ is $\frac{a}{a+b} = \frac{9}{32} \approx 0.28$.
