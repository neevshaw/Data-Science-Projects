```r
ggplot(mpg, aes(displ, hwy, color=class, size=2)) + geom_point(shape=15) + geom_smooth(linewidth=1, method="lm", se=FALSE, color="purple") + 
theme(panel.background = element_rect(fill = "lightgreen"), plot.background = element_rect(fill = "red"),
      plot.title = element_text(size = 30, color="yellow"), legend.text=element_text(size=20),
      legend.title=element_text(size=20), axis.title.x = element_text(size = 18, color="blue"), 
     axis.title.y = element_text(size = 18, color="blue"), axis.text.y = element_text(size = 14, color="white"),
     axis.text.x = element_text(size = 14, color="white")) + 
ggtitle("Christmas MPG") + labs(x="Displacement", y="Highway", color="Type of Car") + scale_size(guide = "none")
```
