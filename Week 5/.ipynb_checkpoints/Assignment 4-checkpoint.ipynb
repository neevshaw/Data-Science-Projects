## Part (e)

We need the hypervolume to be 10% of the total hypervolume which is 1. Thus $s^p = 0.1$ so $s = \sqrt[p]{0.1}$

For $p = 1$, $s = 0.1$

For $p = 2$, $s \approx 0.32$

For $p = 100$, $s \approx 0.98$
