# Question 1

```r
Donner = read.csv("DonnerParty.csv")
attach(Donner)
fix(Donner)
summary(Donner)
```

As introduced in the beginning of the quiz this data refers to the Donner Party - a group of emigrants who faced harsh conditions during the winter of 1846-1847 in the Sierra Nevada Mountains. The columns in this dataset correspond to the following:
* Name: Name of the emigrant
* Age: Age of the emigrant
* Survive: binary variable corresponding to whether or not the emigrant survived the winter (survived = 1)
* sex (lowercase): binary text variable corresponding to sex of the emigrant
* Sex (uppercase): binary variable corresponding to sex (but Male = 1)
* Status: a varaible with 3 classifications correspodning to status of emigrant (Family, Single, Hired)
* Hired: binary variable where Hired = 1 if Status = Hired

## Part (a)

### (i) How many men were in the Donner party?

```r
Male = Donner[Sex==1,]
print(nrow(Male))
```

We mask the Donner dataframe and only lopok at values for which Sex = 1 (i.e. Males). Looking at the number of rows in the masked dataframe gives that that were 53 men in the Donner Party.

### (ii) What was the survival rate for men?

```r
print(sum(Male$Survive)/nrow(Male))
```

If a man survived, their Surive variable is 1 - and if they didn't the Surve variable is 0. Adding up the entier Survive columsn, therefore, tells up how many total men survived. Dividing that by the total amount of men gives a survival rate of 45.28%.

## Part (b)

```r
model_age = glm(Survive~Age, family=binomial)
summary(model_age)
```

We simply take the coefficients provided in the coefficients table to get:

## Part (c)

```r
model_sex = glm(Survive~Sex, family=binomial)
summary(model_sex)
```

Conducting a logistic regression with Sex instead of Age, we see that the P-value for Sex is slightly higher than for Age (0.0174 as compared to 0.0103) meaning Age is a better predictor.

## Part (d)

```r
print(exp(confint(model_sex, parm="Sex")))
```

Creating a 95% confidence interval for the Sex parameter gives it in terms of log-odds, so we apply the $\exp$ function to get the odds ratio.

## Part (e)

Note that the odds ratio is computed as $\dfrac{p}{1 - p}$. The confidence interval is contained within the interval (0, 1) meaning that $p < 1 - p$. Since $p$ is the probability that if Sex = 1 (male), then Survive = 1 (survives), this means that women in the Donner Party are more likely to survive than men.

# Question 2

## Part (a)

```r
model_sat = glm(Survive~Sex+Age+I(Age^2), family=binomial)
summary(model_sat)
```

Again we look at the coefficients table to get:
$$ \ln(\text{odds}) = 1.01255 - 1.221738\cdot\text{Sex} + 0.073064\cdot\text{Age} - 0.002335\cdot\text{Age}^2$$

## Part (b)

```r
model_reduced = glm(Survive~Sex+Age, family=binomial)
anova(model_reduced, model_sat, test="Chisq")
```

The p-value is 0.03, which means we reject the null hypothesis. There is a significant difference between the two models, so the $\text{Age}^2$ term is significant.

## Part (c)

```r
presid = residuals(model_sat, type="pearson")
print(max(abs(presid)))
```

We see that the maximum absolute value of the Pearson residual is 3.719 (which is positive as well in the original dataset)

Conducting a $\chi^2$ test on the contingency table, gives a p-value of 0.178, which means that there is nt a significant difference in the survival rates between hired and not hired emigrants in the Donner Party.
