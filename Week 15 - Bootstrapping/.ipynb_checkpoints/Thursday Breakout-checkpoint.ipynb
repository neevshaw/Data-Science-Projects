```r
housing = read.csv("housing.csv")
head(housing)
attach(housing)
```

```r
library(quantreg)
library(ggplot2)
qr = rq(median_house_value~median_income, data=housing, tau=0.5)
summary(qr)
ggplot(housing, aes(median_income, median_house_value)) + geom_point() + geom_quantile(quantiles = 0.5)
```

```r
qs <- 1:9/10
qr2 <- rq(median_house_value ~ median_income, data=housing, tau = qs)
```

```r
ggplot(housing, aes(median_income, median_house_value)) + geom_point() + geom_quantile(quantiles = qs)
```

```r
summary(qr2)
```

```r
plot(summary(qr2), parm="median_income")
```
