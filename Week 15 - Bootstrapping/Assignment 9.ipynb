## Part (a)

The $k$-fold cross-validation technique is implemented by dividing the data into $k$ groups called folds and then subsequently training whatever model or fit on $k-1$ fold while evaluating it on the held-out fold. We repeat this for all $k$ folds and return the cross-validation estimate as the average of all $k$ test errors.

## Part (b)

### Part (i) Validation set approach

The advatnages of $k$-fold cross-validation over the validation set approach are very apparent. First with $k$ folds we get a better estimate of the actual test error since we are averaging models trained on most of the data set but varying the held-out fold each time. The validation set approach is only trained once while leaving out the validation set entirely leading to a model with less training data and a higher error rate. 

### Part (ii) LOOCV

LOOCV has the opposite problem. First, since we are only leaving out one data point, we are training $n$ models which can be computationally epxensive for large data sets. However, since LOOCV is averaging so many models, it tends to give an unbiased estimate compared to $k$-fold CV. However, on the flip side, LOOCV has a much higher variance since each of the models that are being trained are almost identical to all the other model - with only 1 data point extra and missing between each model. Therefore there is a bias-variance tradeoff when it comes to $k$-fold CV (specifically a low value of $k$ such as 5 or 10) and LOOCV.
