## Exercise 2.1

### Model A:

\begin{align*}
p(1) &= \dfrac14 \\
p(2) &= \dfrac14 \\
p(3) &= \dfrac14 \\
p(4) &= \dfrac14 \\
\end{align*}

This model has no bias and each outcome is equally likely.

### Model B:

\begin{align*}
p(1) &= \dfrac1{10} \\
p(2) &= \dfrac2{10} \\
p(3) &= \dfrac3{10} \\
p(4) &= \dfrac4{10} \\
\end{align*}

This model is biased towards higher numbers, with 4 being the most likely roll.

### Model C:

\begin{align*}
p(1) &= \dfrac{12}{25} \\
p(2) &= \dfrac{12}{50} = \dfrac6{25} \\
p(3) &= \dfrac{12}{75} = \dfrac4{25} \\
p(4) &= \dfrac{12}{100} = \dfrac3{25} \\
\end{align*}

This model is biased towards lower numbers with 1 being the most likely roll.
