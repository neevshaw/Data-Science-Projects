# Exercise 8.5

## Part (a)

The distribution for $y$ is $binomial\left(n=116,\pi=\frac{17}{116}\right)$.

## Part (b)

The posterior distribution is $beta(1+17,4+(116-17)) = beta(18,103)$

## Part (c)

The first moment is the mean $\left(m'\right)$ which is
$$\text{E}[\pi | y] = \dfrac{a}{a+b} = \dfrac{18}{18+103} = \dfrac{18}{121}.$$
The second moment is the variance $\left((s')^2\right)$ which is
$$\text{Var}[\pi | y] = \dfrac{a\times b}{(a + b)^2\times (a + b - 1)} = \dfrac{18\times 103}{(18+103)^2\times (18 + 103 + 1)} = \dfrac{927}{893101}.$$

## Part (d)

The normal approximation is simple $normal\left(m', (s')^2\right) = normal\left(\frac{18}{121}, \frac{927}{893101}\right)$.

## Part (e)

The 95% credible interval of the normal distribution found above is simply $m' \pm 1.96 \times s'$, and plugging in our values above for $m'$ and $s' = \sqrt{\text{Var}[\pi | y]}$, we get $(0.086, 0.212)$.
