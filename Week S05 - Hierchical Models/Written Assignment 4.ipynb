## Part (a)

The frequentist estimator is simply the value of $\pi$ observed in the sample which is $\hat{\pi_f}=\frac{11}{116}\approx0.095$.

## Part (b)

The posterior distribution $g(\pi\mid y)$ is simply $beta(1+11,10+(116-11)) = beta(12, 115)$.

## Part (c)

$$\text{E}[\pi\mid y] = \frac{12}{12+115} \approx 0.094$$
$$\text{Var}[\pi \mid y] = \frac{12\times115}{(12+115)^2(12+115+1)}\approx0.00067$$
The Bayesian estimator is simply the expected value: $\hat{\pi_b} = E[\pi \mid y] \approx 0.094$.

## Part (d)

The 95% credible interval is
$$\text{E}[\pi\mid y] \pm 1.96\times\sqrt{\text{Var}[\pi\mid y]} \approx (0.044,0.145).$$

## Part (e)

Since $\pi = 0.10$ lies within the 95% credible interval above, we fail to reject the null hypothesis.

## Part (a)

```r
pD = function(z,N,w,k) { exp( lbeta(z+w*(k-2)+1,N-z+(1-w)*(k-2)+1) - lbeta(w*(k-2)+1,(1-w)*(k-2)+1) ) }
```

```r
pOdds = (pD(7, 10, 0.25,6) * 0.5)/(pD(7, 10, 0.75,6) * 0.5)
pM1 = pOdds/(1 + pOdds)
pM2 = 1 - pM1
```

```r
cat("Posterior Model 1: ",pM1,"\n")
cat("Posterior Model 2: ",pM2,"\n")
```

A formal analysis reveals that the posterior probabilities for each factory are 0.25 and 0.75 (i.e. there is a 75% probability that the coin came from the second factory given the data).
