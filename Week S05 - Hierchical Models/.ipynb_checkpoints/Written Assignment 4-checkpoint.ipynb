## Part (a)

```r
pD = function(z,N,w,k) { exp( lbeta(z+w*(k-2)+1,N-z+(1-w)*(k-2)+1) - lbeta(w*(k-2)+1,(1-w)*(k-2)+1) ) }
```

```r
pOdds = (pD(7, 10, 0.25,6) * 0.5)/(pD(7, 10, 0.75,6) * 0.5)
pM1 = pOdds/(1 + pOdds)
pM2 = 1 - pM1
```

```r
cat("Posterior Model 1: ",pM1,"\n")
cat("Posterior Model 2: ",pM2,"\n")
```

JAGS gives us an approximate posterior of 0.251 for model 1 and 0.749 for model 2, while a formal analysis reveals that the exact probabilities are indeed 0.25 and 0.75 (i.e. there is a 75% probability that the coin came from the second factory given the data).

## Part (b)
